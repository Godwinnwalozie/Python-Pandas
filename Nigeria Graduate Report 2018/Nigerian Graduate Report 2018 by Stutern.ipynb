```python
# find out the count of male vs female on the survey
nig_grad_gender = nig_grad['What is your gender'].value_counts().to_frame()
nig_grad_gender
```

```python
# what percentage of female is the sample
female_perc = nig_grad_gender.iloc[1,0] / ( nig_grad_gender.iloc[0,0] +\
        nig_grad_gender.iloc[1,0] + nig_grad_gender.iloc[2,0]) * 100
female_perc = round(female_perc,2)
female_perc = print(str(female_perc) + ' %') 
```

```python
# what percentage of male is the sample
male_perc = nig_grad_gender.iloc[0,0] / ( nig_grad_gender.iloc[0,0] +\
        nig_grad_gender.iloc[1,0] + nig_grad_gender.iloc[2,0]) * 100
male_perc = round(male_perc,2)
male_perc = print(str(male_perc) + ' %') 
```

```python
# which year had the largest graduates
nig_grad['Year of graduation'].value_counts().to_frame()\
.sort_values('Year of graduation', ascending = False)
```

```python
# which school had the highest graduate
nig_grad['Polytechnic/University attended'].value_counts().to_frame()
```

```python
# what is the highest level of degree by 
grad_degree = nig_grad.loc[:, ['What is your highest level of education?','What is your gender','Timestamp']].\
groupby(['What is your highest level of education?','What is your gender']).\
count().rename({'Timestamp' : 'count'},axis = 1)
```
