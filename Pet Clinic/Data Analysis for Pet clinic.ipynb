# Question 1 
# Extract information on Pets names and owner names side by side

```python
# Question 1 
# Extract information on Pets names and owner names side by side
# step 1 , merge 'pet_data'  and pet_owner' 
pet_owner = pd.merge(pet_data,pet_owner, left_on = 'OwnerID',\
                     right_on ='OwnerID', how = 'inner')
```

```python
pet_owner.head()
```

```python
pet_owner1 = pet_owner.loc[:, ['Name_x', 'Kind','Name_y','Surname']].rename({'Name':'Pet_name'}, axis = 1)
```

```python
pet_owner1['Owner_name'] = pet_owner1.Name_y + '  ' + pet_owner1.Surname
```

```python
pet_owner1.loc[:, ['Name_x','Kind','Owner_name']].drop_duplicates\
(subset = ['Name_x','Kind','Owner_name'], keep = 'first')
```

# Question 2
# Which Pet from the clinic had procedures performed?

```python
# lets merge 'pet_data' and 'pet_procedures_history'
pet_data_hist = pd.merge(  pet_data, pet_procedures_history,left_on =\
                        'PetID',right_on ='PetID',how= 'inner' )
```

```python
pet_has_procedure = pd.merge( pet_data_hist, pet_procedures_detail, left_on= \
         'ProcedureType', right_on = 'ProcedureType',how= 'inner')
```

```python
pet_has_procedure[pet_has_procedure.ProcedureType.notna()].loc[:, \
['Name','Kind','ProcedureType','Name']].drop_duplicates(subset=['Name','Kind','ProcedureType'], keep ='first')
```

# Question 3
# Match all the procedures perfomed to their description

```python
pet_has_procedure.loc[:, ['ProcedureType','Description']].\
drop_duplicates(subset = ['ProcedureType','Description'], keep = 'first')
```
