```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame( { 'name' : ['Peter', 'James' , 'Mariam', 'Godwin', 'Annie', 'Laura'],\
                    'Age': [34, 54, 21, 14, 29,30], 'Score' : [56,46,30,50,100,30]})
```

```python
df
```

```python
# add a new colum and divide score by age
df['Diff'] = round(df.Score/df.Age,2)
```

```python
df
```

```python
# add 10 to the score mark
df.Score = df.Score.apply(lambda x: x + 10)
```

```python
df
```

```python
# add 10 to the score mark if score is greater 40
df.Score = df.Score.apply(lambda x: x-10 if x > 40 else x)
```

```python
# add 100 to scores
df.Score = df.Score.apply (lambda x: x +1000 if x == 40 else x)
```

```python
df['Total'] = df['Age'] + df['Score']
```
